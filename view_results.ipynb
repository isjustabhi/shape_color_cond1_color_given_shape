```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Paths to results
samples_dir = "samples/cond1"
epochs = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]

# Plot all grayscale inputs and color outputs
plt.figure(figsize=(12, len(epochs) * 2))

for idx, epoch in enumerate(epochs):
    gray_path = os.path.join(samples_dir, f"input_gray_epoch{epoch}.png")
    color_path = os.path.join(samples_dir, f"output_color_epoch{epoch}.png")

    if not os.path.exists(gray_path) or not os.path.exists(color_path):
        print(f"Missing files for epoch {epoch}, skipping...")
        continue

    gray_img = Image.open(gray_path).convert("L")
    color_img = Image.open(color_path)

    # Show grayscale
    plt.subplot(len(epochs), 2, 2 * idx + 1)
    plt.imshow(gray_img, cmap="gray")
    plt.title(f"Gray Input - Epoch {epoch}")
    plt.axis("off")

    # Show color
    plt.subplot(len(epochs), 2, 2 * idx + 2)
    plt.imshow(color_img)
    plt.title(f"Color Output - Epoch {epoch}")
    plt.axis("off")

plt.tight_layout()
plt.savefig("samples/cond1/all_epoch_comparisons.png")
plt.show()
```
