```python
# scripts/view_results.ipynb (for conditional model)

import os
from PIL import Image
import matplotlib.pyplot as plt

# Folder where model output samples are stored
samples_dir = "samples/cond1"

# Find available epoch numbers from saved images
epochs = sorted([
    int(f.split("epoch")[1].split(".")[0])
    for f in os.listdir(samples_dir)
    if f.startswith("output_color_epoch")
])

# Function to display grayscale input and predicted color output
def show_epoch(epoch):
    gray_path = os.path.join(samples_dir, f"input_gray_epoch{epoch}.png")
    color_path = os.path.join(samples_dir, f"output_color_epoch{epoch}.png")

    if os.path.exists(gray_path) and os.path.exists(color_path):
        gray_img = Image.open(gray_path)
        color_img = Image.open(color_path)

        plt.figure(figsize=(10, 5))

        plt.subplot(1, 2, 1)
        plt.title(f"Input (Gray)  Epoch {epoch}")
        plt.imshow(gray_img, cmap='gray')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.title("Predicted Color Output")
        plt.imshow(color_img)
        plt.axis('off')

        plt.tight_layout()
        plt.show()
    else:
        print(f"Missing files for epoch {epoch}.")

# Show latest epoch result
show_epoch(epochs[-1])
```
